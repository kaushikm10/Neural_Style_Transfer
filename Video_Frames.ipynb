```python
import os
import cv2
import numpy as np
import glob
img_array = []
for filename in glob.glob('new_frames/*.jpg'):
    print(filename)
    img = cv2.imread(filename)
    h, w, l = img.shape
    size = (w, h)
    img_array.append(img)
out = cv2.VideoWriter('style_transfer_movie.avi', cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
```

```python
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
